```python
class derivative () :
    def __init__ (self, I) :
        self.I = I
    
    def compute_1 (self) :

        return J_x, J_y

    def compute_2 (self) :

        J_xx, J_yy, J_xy

class Image (derivative) :
    def __init__ (self, I) :
        self.I = I
        self.m, self.m = np.shape (self.I)


    def __call__ (self,) :
        return imshow (image, cmap = "gray")
    

    def neumann_boundary (self, border_size = 20) :
        d = border_size

        M = self.m + 2 * d
        N = self.n + 2 * d

        J = np.zeros ((M, N))
        J [(d-1):(M-d-1), (d-1):(N-d-1)] = I

        for i in range (self.m) :
            for j in range (d) :
                J [i+d-1, j] = self.I [i, d-j-1]
            
            for j in range (N-d-1, N) :
                J[i+d-1,j] = self.I [i, N-d-j-1]

        for j in range (N) :
            for i in range (d) :
                J [i,j] = J [2*d-1-i, j]
            
            for i in range (M-d-1, M) :
                J [i,j] = J [2*(M-d-1)-1-i, j]

        return J


    def gaussian_noise (self, noise = 5) :
        self.m, self.n = np.shape (self.I)
        
        J = np.zeros ((self.m,self.n))
        J = noise * np.random.randn(self.m)
        
        return self.I + J


    def heat_equation (self, dt = .25, noise = 5, N = 100) :
        I0 = gaussian_noise ()
        
        J = I0
        for i in range (N) :
            J_xx, J_yy, J_xy = derivative.compute_2 (J)
            
            laplacien_J = J_xx + J_yy
            J += laplacien_J * dt
        
        return J
```
